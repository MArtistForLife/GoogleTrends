## Handling Imbalanced Dataset

1. Up Sampling - increase data points from the minority category
2. Down Sampling - decrease data points from the majority category

```python
import numpy as np
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)
# when randomly selecting values, it won't change

# Create a dataframe with two classes
n_samples = 1000 # take 1000 data points
class_0_ratio = 0.9 # a 9 to 1 ratio
n_class_0 = int(n_samples * class_0_ratio) # number of samples x that ratio
n_class_1 = n_samples - n_class_0 # the leftover (minority in this case)
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})
# loc = 0 means that the mean of distribution is 0
# scale = 1 means the std of distribution is 1
# target says that we will have a list of length n_class_0 (in this case, 900)
# just filled with zeroes
# all rows in this df belong to class 0

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
# loc = 2 means that the mean of distribution is 2
# scale = 1 means the std of distribution is still 1
# target says that we will have a list of length n_class_1 (in this case, 100)
# just filled with ones
# all rows in this df belong to class 1
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
# concatenate both df to make one single dataset
```

```python
df.head()
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

```python
## upsampling - increase the minority!!
df_minority=df[df['target']==1] # where the target is 1 (the minority 100)
df_majority=df[df['target']==0] # where the target is 0 (the majority 900)
```

```python
from sklearn.utils import resample
# resample creates more new points for minority and try to
# equalize with respect to majority

df_minority_upsampled=resample(df_minority, replace=True, n_samples=len(df_majority), random_state=42)
# replace = True means the sample will get added as replacement (here, for upsampling)
# n_samples = len(df_majority) bc we want to increase by as much as the majority
# you need random_state to make sure your outputs are the same every time AND that the rows aren't randomly
# selected with replacement (42 is just a commonly used seed value bc of
# Hitchhiker's Guide to the Galaxy)

# note: random_state will keep your randomized outputs the same during THIS session
# (they will always turn out something different if this session disconnects and
# you rerun everything in a new session)
# without random_state, every time you run this block of CODE (during one session)
# you will get different outputs, which makes it harder to troubleshoot for errors
```

```python
df_minority_upsampled.shape
# now we have as many minority points as majority points
```

```python
df_minority_upsampled.head()
# these all belong to the minority category!!
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
# now we have equalized the dataset with upscaling
```

```python
df_upsampled['target'].value_counts()
# total is now 1800
```

## Down Sampling
##### Same kind of code involved, just a little tweaking between the minority and majority category variables.

```python
import pandas as pd

# Set the random seed for reproducibility
np.random.seed(123)

# Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})

df = pd.concat([class_0, class_1]).reset_index(drop=True)

# Check the class distribution
print(df['target'].value_counts())
```

```python
## downsampling - decrease the majority!!
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
from sklearn.utils import resample
df_majority_downsampled=resample(df_majority,replace=True, n_samples=len(df_minority), random_state=42)
# same idea here, but now with the majority category involved
```

```python
df_majority_downsampled.shape
# ta-dah :)
```

```python
df_downsampled=pd.concat([df_minority,df_majority_downsampled])
df_downsampled.head()
```

```python
df_downsampled['target'].value_counts()
```
